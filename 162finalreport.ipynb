## Abstract

Attention-Deficit/Hyperactivity Disorder (ADHD) is associated with alterations in large-scale brain connectivity patterns. In this project, we investigate whether a transformer-based classifier can distinguish ADHD patients from healthy controls using resting-state functional MRI (rs-fMRI) data. Functional connectivity matrices are computed from brain region time series and used as inputs to a transformer model that captures relationships between brain regions through self-attention mechanisms. We evaluate classification performance using metrics such as accuracy, ROC-AUC, and F1-score, and compare the transformer model to traditional machine learning baselines. Understanding which connectivity patterns contribute most strongly to classification may provide insights into the network-level neural differences associated with ADHD.

## Research Question

Can a transformer-based classifier distinguish ADHD patients from healthy controls using resting-state functional MRI (rs-fMRI) by learning differences in brain functional connectivity patterns?

## Introduction and Background

Attention-Deficit/Hyperactivity Disorder (ADHD) is a common neurodevelopmental condition characterized by inattention, impulsivity, and executive dysfunction (Kuntsi et al., 2006). Neuroimaging studies have shown that ADHD is associated with altered functional connectivity across several large-scale brain networks, including the default mode network, frontoparietal control network, and attention networks (Silk et al., 2005). Resting-state functional MRI (rs-fMRI) provides a noninvasive way to measure spontaneous brain activity and has been widely used to investigate connectivity abnormalities in ADHD. However, identifying reliable biomarkers from rs-fMRI remains challenging due to noise, inter-subject variability, and high dimensionality of brain imaging data.

Recent advances in deep learning, particularly transformer architectures, have demonstrated strong performance in modeling complex relationships in high-dimensional data. Transformers are well suited for functional connectivity analysis because they can capture long-range dependencies between brain regions through self-attention. Prior work applying transformer models to resting-state fMRI has shown improved classification performance compared to traditional machine learning approaches such as support vector machines and multilayer perceptrons (Dai et al., 2023). In this project, we apply a transformer-based classifier to rs-fMRI data to determine whether connectivity patterns can reliably distinguish ADHD patients from healthy controls, and to explore which brain regions contribute most to classification.

The dataset used in this project comes from the ADHD-200 dataset, a widely used neuroimaging dataset designed to support the development of computational methods for ADHD diagnosis. The dataset contains resting-state fMRI recordings from both ADHD patients and healthy control subjects across multiple research sites. Each scan captures blood-oxygen-level-dependent (BOLD) signals across the entire brain over time, allowing researchers to study patterns of functional connectivity between brain regions. By computing correlations between the time series of different brain regions, we can construct functional connectivity matrices that represent the brain as a network of interacting regions.

This project builds upon recent work that applied transformer architectures to large-scale resting-state fMRI datasets for psychiatric disorder classification. In particular, Dai et al. (2023) demonstrated that transformer models can outperform traditional machine learning approaches such as support vector machines when modeling complex connectivity relationships in neuroimaging data. Inspired by this work, we apply a transformer-based classifier to the ADHD-200 dataset to determine whether similar architectures can capture connectivity patterns that differentiate ADHD patients from healthy controls. This approach also allows us to explore which brain regions contribute most strongly to classification through the model’s attention mechanisms.

## Dataset

We use resting-state fMRI data from two groups of subjects. ADHD patients correspond to subject IDs 70001–70056 (42 total), and healthy controls correspond to subject IDs 10159–11156 (127 total). Each subject has a resting-state BOLD fMRI scan stored in NIfTI format. The dataset contains whole-brain time-series data that allow estimation of functional connectivity between brain regions.

## Data Preprocessing

DPARSF software (Yan & Zang, 2010) is used for pre-processing fMRI data (http://www.rfmri.org/DPARSF). The pre-processing of the data follows the procedure of (Yan et al., 2019), including: the first 10 volumes of the BOLD signal were discarded, slice-timing correction was used to ensure the temporal consistency of images across all layers within a single scanning session. T1-weighted images were co-registered to the mean functional image using a 6 degrees-of-freedom linear transformation and then segmented into gray matter (GM), white matter (WM), and cerebrospinal fluid (CSF). At last, the signals were transformed to MNI space and the time series were filtered through a temporal bandpass filter (0.01–0.1Hz).

Raw fMRI scans are converted into functional connectivity representations. Brain activity is parcellated into regions of interest using a predefined brain atlas, producing regional time-series data for each subject. Pearson correlation is then computed between all pairs of ROI time series to generate a functional connectivity matrix representing the subject’s brain network.

## Feature Extraction 

First, for each subject, brain regions were partitioned using a brain atlas, and the average time series for each brain region was calculated. The atlases used in this study included the AAL-116 atlas and the CC200 atlas (Craddock et al., 2012), which yielded time series signals for 116 and 200 brain regions, respectively. Then, z-score normalization was applied to each time series, which may help alleviate the impact of site differences on the experiment. The features used in this study were FCs between brain regions calculated by Nilearn (https://nilearn.github.io/stable/index.html) using Pearson correlation. The FC matrix obtained using the AAL-116 atlas had a dimension of 116 × 116, while the one obtained using the CC200 atlas was 200 × 200.


## Data Analysis & Results

The model used in this study is the Transformer-Encoder, whose core is self-attention mechanism. Unlike the simple single-attention function, Transformer employs a multiheaded self-attention mechanism. Then, the feed-forward network (FNN) applies two non-linear transformations to the output of the previous layer using activation functions. Finally, to address the issue that deep neural network models are difficult to train, both the multihead self-attention network and the FNN are equipped with a residual connection layer with layer normalization.

First, we extract time series data from rs-fMRI using brain atlases and compute the FC matrix 
 using Pearson correlation. After linear projection, each feature is projected into a feature space of dimension d, and then encoded by position and fed into the encoder of the Transformer. Then we obtain the final latent representation of each subject.
. Finally, the classifier composed of linear layers is applied to obtain the final classification results.

This method is appropriate for our dataset because functional connectivity matrices are high-dimensional and contain complex interactions across brain regions that are difficult for traditional linear models to capture. Attention mechanisms also provide interpretability by highlighting which brain regions contribute most to classification.


## Evaluation

The dataset will be split into training and validation sets using stratified sampling to preserve class balance. Model performance will be evaluated using classification accuracy, ROC-AUC, and F1-score. We will also compare the transformer classifier against a baseline model (e.g., logistic regression or support vector machine) trained on the same functional connectivity features. Finally, we will analyze attention weights or gradient-based importance scores to identify brain regions that most strongly influence ADHD classification.
