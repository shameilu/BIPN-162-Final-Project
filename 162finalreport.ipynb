## Abstract

Attention-Deficit/Hyperactivity Disorder (ADHD) is associated with alterations in large-scale brain connectivity patterns. In this project, we investigate whether a transformer-based classifier can distinguish ADHD patients from healthy controls using resting-state functional MRI (rs-fMRI) data. Functional connectivity matrices are computed from brain region time series and used as inputs to a transformer model that captures relationships between brain regions through self-attention mechanisms. We evaluate classification performance using metrics such as accuracy, ROC-AUC, and F1-score, and compare the transformer model to traditional machine learning baselines. Understanding which connectivity patterns contribute most strongly to classification may provide insights into the network-level neural differences associated with ADHD.

## Research Question

Can a transformer-based classifier distinguish ADHD patients from healthy controls using resting-state functional MRI (rs-fMRI) by learning differences in brain functional connectivity patterns?

## Introduction and Background

Attention-Deficit/Hyperactivity Disorder (ADHD) is a common neurodevelopmental condition characterized by inattention, impulsivity, and executive dysfunction (Kuntsi et al., 2006). Neuroimaging studies have shown that ADHD is associated with altered functional connectivity across several large-scale brain networks, including the default mode network, frontoparietal control network, and attention networks (Silk et al., 2005). Resting-state functional MRI (rs-fMRI) provides a noninvasive way to measure spontaneous brain activity and has been widely used to investigate connectivity abnormalities in ADHD. However, identifying reliable biomarkers from rs-fMRI remains challenging due to noise, inter-subject variability, and high dimensionality of brain imaging data.

Recent advances in deep learning, particularly transformer architectures, have demonstrated strong performance in modeling complex relationships in high-dimensional data. Transformers are well suited for functional connectivity analysis because they can capture long-range dependencies between brain regions through self-attention. Prior work applying transformer models to resting-state fMRI has shown improved classification performance compared to traditional machine learning approaches such as support vector machines and multilayer perceptrons (Dai et al., 2023). In this project, we apply a transformer-based classifier to rs-fMRI data to determine whether connectivity patterns can reliably distinguish ADHD patients from healthy controls, and to explore which brain regions contribute most to classification.

The dataset used in this project comes from the ADHD-200 dataset, a widely used neuroimaging dataset designed to support the development of computational methods for ADHD diagnosis. The dataset contains resting-state fMRI recordings from both ADHD patients and healthy control subjects across multiple research sites. Each scan captures blood-oxygen-level-dependent (BOLD) signals across the entire brain over time, allowing researchers to study patterns of functional connectivity between brain regions. By computing correlations between the time series of different brain regions, we can construct functional connectivity matrices that represent the brain as a network of interacting regions.

This project builds upon recent work that applied transformer architectures to large-scale resting-state fMRI datasets for psychiatric disorder classification. In particular, Dai et al. (2023) demonstrated that transformer models can outperform traditional machine learning approaches such as support vector machines when modeling complex connectivity relationships in neuroimaging data. Inspired by this work, we apply a transformer-based classifier to the ADHD-200 dataset to determine whether similar architectures can capture connectivity patterns that differentiate ADHD patients from healthy controls. This approach also allows us to explore which brain regions contribute most strongly to classification through the model’s attention mechanisms.
