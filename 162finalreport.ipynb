## Abstract

Attention-Deficit/Hyperactivity Disorder (ADHD) has been associated with disruptions in large-scale brain connectivity networks. In this project, we investigate whether a transformer-based classifier can distinguish ADHD patients from healthy controls using resting-state functional MRI (rs-fMRI) data. Functional connectivity matrices are computed from regional brain time series and used as inputs to a transformer architecture that models interactions between brain regions through multi-head self-attention. We replicate and adapt the transformer-encoder framework proposed by Dai et al. (2023) and apply it to an ADHD rs-fMRI dataset to evaluate whether the model can capture connectivity patterns that differentiate ADHD from typical brain activity. Model performance is assessed using classification metrics including accuracy, ROC-AUC, and F1-score, and results are compared with baseline machine learning approaches.

## Research Question

Can a transformer-based classifier distinguish ADHD patients from healthy controls using resting-state functional MRI (rs-fMRI) by learning differences in brain functional connectivity patterns?

## Introduction and Background

Attention-Deficit/Hyperactivity Disorder (ADHD) is a common neurodevelopmental condition characterized by inattention, impulsivity, and executive dysfunction (Kuntsi et al., 2006). Neuroimaging studies have shown that ADHD is associated with altered functional connectivity across several large-scale brain networks, including the default mode network, frontoparietal control network, and attention networks (Silk et al., 2005). These network-level differences suggest that ADHD may involve disruptions in communication between distributed brain regions rather than localized abnormalities.

Resting-state functional MRI (rs-fMRI) provides a noninvasive method for measuring spontaneous brain activity and has been widely used to investigate connectivity patterns associated with neurological and psychiatric conditions. By examining correlations between time series recorded from different brain regions, researchers can estimate functional connectivity and characterize the brain as a network of interacting regions. However, identifying reliable biomarkers from rs-fMRI remains challenging due to noise, inter-subject variability, and the high dimensionality of neuroimaging data.

Recent advances in deep learning have introduced new approaches for modeling complex relationships in high-dimensional datasets. Transformer architectures, originally developed for natural language processing, have demonstrated strong performance in tasks involving structured and relational data. Because transformers use self-attention mechanisms to model relationships between elements in a sequence, they are well suited for functional connectivity analysis where interactions between distant brain regions are important. In this project, we investigate whether a transformer-based classifier can distinguish ADHD patients from healthy controls using rs-fMRI connectivity patterns.

## Prior/Related Work 

Machine learning methods have increasingly been applied to resting-state fMRI data in an effort to identify biomarkers for ADHD and other neuropsychiatric conditions. Early approaches commonly used traditional classifiers such as support vector machines, logistic regression, and multilayer perceptrons trained on functional connectivity features. For example, the ADHD-200 Global Competition demonstrated that machine learning models could achieve moderate classification performance using neuroimaging features, although performance was often limited by dataset variability and feature engineering challenges (Brown et al., 2012).

More recently, deep learning approaches have been introduced to better capture the complex structure of brain connectivity data. Convolutional neural networks and graph-based models have been applied to functional connectivity matrices to learn representations of brain networks directly from neuroimaging data. These models can capture nonlinear relationships between brain regions and have shown improved performance compared to traditional machine learning methods.

Transformer architectures have recently emerged as a promising approach for analyzing neuroimaging connectivity patterns. Transformers rely on a self-attention mechanism that allows the model to learn relationships between all elements in a sequence simultaneously. Dai et al. (2023) demonstrated that a transformer-encoder model applied to large-scale resting-state fMRI data could successfully classify Major Depressive Disorder by learning connectivity patterns between brain regions. Their work showed that transformer-based models can outperform several traditional machine learning approaches when applied to functional connectivity data.

Building on this work, our project applies a similar transformer-based architecture to classify ADHD using resting-state fMRI data. By replicating and adapting the transformer encoder framework, we evaluate whether attention-based models can effectively capture connectivity patterns that differentiate ADHD patients from healthy controls.

## Dataset

The dataset used in this study comes from the ADHD-200 dataset, a widely used neuroimaging dataset designed to support the development of computational methods for ADHD diagnosis. The dataset contains resting-state fMRI recordings from both ADHD patients and healthy control subjects collected across multiple research sites.

Each subject’s scan contains blood-oxygen-level-dependent (BOLD) signals measured across the entire brain over time. These signals reflect fluctuations in neural activity and can be used to estimate functional connectivity between brain regions. After preprocessing, the brain is parcellated into regions of interest (ROIs) using a predefined brain atlas, producing a set of time series representing activity in each region.

Functional connectivity matrices are computed by calculating Pearson correlation coefficients between the time series of each pair of brain regions. These matrices represent the strength of connectivity between regions and serve as the input features for our classification model.

## Data Preprocessing

DPARSF software (Yan & Zang, 2010) is used for pre-processing fMRI data (http://www.rfmri.org/DPARSF). The pre-processing of the data follows the procedure of (Yan et al., 2019), including: the first 10 volumes of the BOLD signal were discarded, slice-timing correction was used to ensure the temporal consistency of images across all layers within a single scanning session. T1-weighted images were co-registered to the mean functional image using a 6 degrees-of-freedom linear transformation and then segmented into gray matter (GM), white matter (WM), and cerebrospinal fluid (CSF). At last, the signals were transformed to MNI space and the time series were filtered through a temporal bandpass filter (0.01–0.1Hz).

Raw fMRI scans are converted into functional connectivity representations. Brain activity is parcellated into regions of interest using a predefined brain atlas, producing regional time-series data for each subject. Pearson correlation is then computed between all pairs of ROI time series to generate a functional connectivity matrix representing the subject’s brain network.

## Feature Extraction 

First, for each subject, brain regions were partitioned using a brain atlas, and the average time series for each brain region was calculated. The atlases used in this study included the AAL-116 atlas and the CC200 atlas (Craddock et al., 2012), which yielded time series signals for 116 and 200 brain regions, respectively. Then, z-score normalization was applied to each time series, which may help alleviate the impact of site differences on the experiment. The features used in this study were FCs between brain regions calculated by Nilearn (https://nilearn.github.io/stable/index.html) using Pearson correlation. The FC matrix obtained using the AAL-116 atlas had a dimension of 116 × 116, while the one obtained using the CC200 atlas was 200 × 200.


## Data Analysis & Results

The model used in this study is the Transformer-Encoder, whose core is self-attention mechanism. Unlike the simple single-attention function, Transformer employs a multiheaded self-attention mechanism. Then, the feed-forward network (FNN) applies two non-linear transformations to the output of the previous layer using activation functions. Finally, to address the issue that deep neural network models are difficult to train, both the multihead self-attention network and the FNN are equipped with a residual connection layer with layer normalization.

. Finally, the classifier composed of linear layers is applied to obtain the final classification results.

This method is appropriate for our dataset because functional connectivity matrices are high-dimensional and contain complex interactions across brain regions that are difficult for traditional linear models to capture. Attention mechanisms also provide interpretability by highlighting which brain regions contribute most to classification.


## Evaluation

The dataset will be split into training and validation sets using stratified sampling to preserve class balance. Model performance will be evaluated using classification accuracy, ROC-AUC, and F1-score. We will also compare the transformer classifier against a baseline model (e.g., logistic regression or support vector machine) trained on the same functional connectivity features. Finally, we will analyze attention weights or gradient-based importance scores to identify brain regions that most strongly influence ADHD classification.
